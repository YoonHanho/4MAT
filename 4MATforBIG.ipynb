#### 데이터 전처리

```python
# 4MAT 설문결과값 입력(N=10)
mat = [['HH',44,40,36,30,-9],
       ['SY',39,38,49,24,9],
       ['DH',18,38,43,51,7],
       ['IH',31,46,38,35,-5],
       ['KN',36,45,29,40,-3],
       ['HD',23,44,33,51,-11],
       ['HC',22,51,44,33,-5],
       ['DB',33,28,33,56,11],
       ['JS',38,48,41,23,-3],
       ['RS',33,36,45,36,-9]]
df_mat = pd.DataFrame(mat)
df_mat.columns = ['이름','동그라미','네모','세모','별','doing_score']
```

```python
# 검산 : 점수합계가 각각 150되는지
df_mat.iloc[:,1:5].sum(axis=1)  # => OK
```

```python
# 피처 추가 : 개인별 최고점수에 대표 세그먼트 넣기
arr = df_mat.iloc[:,1:5].values  # 점수만 array로
max_pos = np.argmax(arr, axis=1) # 최대값 자릿수 찾기
cols = df_mat.columns[1:5] 
df_mat['segment'] = pd.Series([ cols[i] for i in max_pos ])

# 피처 추가 : 점수별 표준편차(4개 세그먼트별 점수가 고르게 분포 되었으면 표준편차가 작다)
df_mat['std'] = pd.Series(np.std(arr, axis=1))
df_mat
```

---
#### 세그먼트 분석 : 건수 집계, 세그먼트별 점수 총점/분포(박스플롯), 상관계수

```python
# 10명의 세그먼트 집계
sns.countplot(x='segment', data=df_mat)
```

네모(과학자형)와 별(창업가형)이 가장 많다.  
반면 동그라미(시인형)은 딱 한명!

```python
# 점수간의 상관분석
corr = df_mat.iloc[:,1:5].corr()

plt.figure(figsize=(8,6))
ax = sns.heatmap(corr, annot=True, linewidths=0.5)
```

별(창업가형) 점수는 다른 세그먼트와 강한 음의 상관관계를 보인다.  
이는 별 유형이 다른 유형과는 상당히 틀린 성향을 보인다는 것을 알 수 있다. 

```python
# 페어플랏 : 관측치가 적어 크게 유효하지는 않음. 분포 확인은 유용
sns.pairplot(df_mat, 
             vars=['동그라미','네모','세모','별'], 
             hue='segment', 
             diag_kind='kde',
             palette='husl')
```

10명의 설문점수를 산점도 행렬로 그려보았다.  
샘플수가 적어 패턴이 잘 보이지 않지만, 파란색의 별 유형과 세모 유형이 튀는 것을 볼 수 있다.  
참고로 샘플수가 1개인 동그라미는 분포가 그려지지 않았다. 똑똑한 seaborn 라이브러리~!

```python
# 밀도함수로 그려보자 : https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot
fig, axes = plt.subplots(2,3, figsize=(10,10))
plt.subplots_adjust(hspace=0.3)

sns.kdeplot(df_mat['동그라미'].values, df_mat['세모'].values, shade=True, ax=axes[0,0]).set_title('동그라미 X 세모')
sns.kdeplot(df_mat['동그라미'].values, df_mat['네모'].values, shade=True, ax=axes[0,1]).set_title('동그라미 X 네모')
sns.kdeplot(df_mat['동그라미'].values, df_mat['별'].values, shade=True, ax=axes[0,2]).set_title('동그라미 X 별')
sns.kdeplot(df_mat['세모'].values, df_mat['네모'].values, shade=True, ax=axes[1,0]).set_title('세모 X 네모')
sns.kdeplot(df_mat['세모'].values, df_mat['별'].values, shade=True, ax=axes[1,1]).set_title('세모 X 별')
sns.kdeplot(df_mat['네모'].values, df_mat['별'].values, shade=True, ax=axes[1,2]).set_title('네모 X 별')
```

4개 유형별 상관관계는 밀도함수 차트가 명확한 패턴을 보여준다. 샘플수가 적을 때 꼭 그려보자.   
별과 만나면 무조건 강한 음의 상관관계를 보인다.

#### 개인별 분석 : 레이더 차트, 표준편차 분석, 다차원척도법(MDS),  클러스터링맵

```python
# 레이더 차트(코드소스) : https://python-graph-gallery.com/392-use-faceting-for-radar-chart/
from math import pi
 
# 소스코드의 데이터 구성은 이렇다. 
# 동일하게 맞춰야 차트 함수 구현됨
'''
df = pd.DataFrame({
'group': ['A','B','C','D'],
'var1': [38, 1.5, 30, 4],
'var2': [29, 10, 9, 34],
'var3': [8, 39, 23, 24],
'var4': [7, 31, 33, 14],
'var5': [28, 15, 32, 14]
})
'''

def make_spider( row, title, color): 
  # number of variable
  categories=list(df)[1:]
  N = len(categories)

  # What will be the angle of each axis in the plot? (we divide the plot / number of variable)
  angles = [n / float(N) * 2 * pi for n in range(N)]
  angles += angles[:1]

  # Initialise the spider plot
  # ax = plt.subplot(2,2,row+1, polar=True, )
  ax = plt.subplot(5,2,row+1, polar=True, )

  # If you want the first axis to be on top:
  ax.set_theta_offset(pi / 2)
  ax.set_theta_direction(-1)

  # Draw one axe per variable + add labels labels yet
  plt.xticks(angles[:-1], categories, color='grey', size=8)

  # Draw ylabels
  ax.set_rlabel_position(0)
  #plt.yticks([10,20,30], ["10","20","30"], color="grey", size=7)
  #plt.ylim(0,40)
  plt.yticks([10,20,30,40,50,60], ["10","20","30","40","50","60"], color="grey", size=7)
  plt.ylim(0,60)

  # Ind1
  values=df.loc[row].drop('group').values.flatten().tolist()
  values += values[:1]
  ax.plot(angles, values, color=color, linewidth=2, linestyle='solid')
  ax.fill(angles, values, color=color, alpha=0.4)

  # Add a title
  plt.title(title, size=11, color=color, y=1.1)
```

```python
# 개인별 차트를 돌려보자
df = df_mat.iloc[:,:5].rename(columns={'이름':'group'}) 

my_dpi=96   # 해상도
#plt.figure(figsize=(1000/my_dpi, 1000/my_dpi), dpi=my_dpi)
plt.figure(figsize=(8,25), dpi=my_dpi)

# Create a color palette:
my_palette = plt.cm.get_cmap("Set2", len(df.index))

# Loop to plot
for row in range(0, len(df.index)):
  # make_spider( row=row, title='group '+df['group'][row], color=my_palette(row))
  make_spider( row=row, title=df['group'][row], color=my_palette(row))
```

#### 다차원 척도법으로 개인간의 유사도를 시각화해보자

4개 유형 점수로 유사도를 계산하고 2차원으로 데이터를 축소하여 차트로 구현  
PCA와 함께 다차원 데이터를 2차원으로 시각화시 적극적으로 활용해보자  
다차원척도법 참고 : https://goo.gl/nHHKNG  

```python
# 정규분포로 스케일 조정
from sklearn.preprocessing import StandardScaler    # 피처별로 스케일 다른 경우 반드시 표준화한다. 
from sklearn import manifold                        # 사이킷런의 manifold를 쓰자~! 

X = df_mat.iloc[:,1:5].values
targets = df_mat.iloc[:,0].values

X_norm = StandardScaler().fit(X).transform(X)

''' 
1. 개체간의 거리행렬, 상관계수 등이 이미 구해져 있는 경우 : dissimilarity="precomputed"
2. 원본 데이터에서 거리를 구해야 하는 경우 : dissimilarity="euclidean"
'''
mds = manifold.MDS(n_components=2, max_iter=1000, eps=1e-9, random_state=40,
                   dissimilarity="euclidean")  
pos = mds.fit(X_norm).embedding_


# 텍스트 산점도 차트 구현
xmax, ymax = np.amax(pos, axis=0)  # 차트 크기 기준값 뽑기
xmin, ymin = np.amin(pos, axis=0)

plt.figure(figsize=(8,8))
for i in range(len(pos)):
  x = pos[i,0]
  y = pos[i,1]
  t = targets[i] 
  plt.text(x, y, t, fontsize=14)
  
# 최대값, 최소값에서 0.1씩 추가해서 공간 더 마련하기
plt.xlim(xmin-0.1, xmax+0.1)
plt.ylim(ymin-0.1, ymax+0.1)
plt.show()
```

상단 : 별 유형이 강하게 응집(DH, HD, DB)  
우하단 : 세모, 네모 유형 응집(RS, JS, SY)
